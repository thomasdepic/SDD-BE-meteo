## Getting an inventory of csv available

```python
# hardcoded directories
data_dir = "./data/"
mesures_dir =  data_dir + "mesures/"

# get a list of csv files
station_csv = data_dir + "stations.csv"
mesures_csvs = [mesures_dir + mes for mes in os.listdir(mesures_dir)]
n_mesures = len(mesures_csvs)
```
